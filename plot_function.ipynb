# 如何画一个图像

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

## 定义一个函数

```python
def f(x):
    return x**x / (2*x + 1)
```

## 画出这个函数

```python
x = np.linspace(0, 2, 1000)
y = f(x)
```

```python
plt.figure(figsize=(10, 8))
plt.plot(x, y)
```

```python
A = np.array([[1, 0, 0], [0, 5, 2],  [3, 1, 9]])
A
```

```python
A_inv = np.linalg.inv(A)
A_inv
```

```python
np.dot(A, A_inv)
```

```python
eigenvalues, eigenvectors = np.linalg.eig(A)
```

```python
print(eigenvalues)
print(eigenvectors)
```

```python
A = np.array([[1, 0, 0], [0, 5, 2], [3, 1, 9]])
v1 = A[:, 0]  # 第一列向量 [1, 0, 3]
v2 = A[:, 1]  # 第二列向量 [0, 5, 1]
v3 = A[:, 2]  # 第三列向量 [0, 2, 9]

fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# 绘制三个向量，使用不同颜色和标签
ax.quiver(0, 0, 0, v1[0], v1[1], v1[2], color='r', label='v1 = [1, 0, 3]', arrow_length_ratio=0.1)
ax.quiver(0, 0, 0, v2[0], v2[1], v2[2], color='g', label='v2 = [0, 5, 1]', arrow_length_ratio=0.1)
ax.quiver(0, 0, 0, v3[0], v3[1], v3[2], color='b', label='v3 = [0, 2, 9]', arrow_length_ratio=0.1)

# 设置坐标轴范围和标签
ax.set_xlim([0, 10])
ax.set_ylim([0, 10])
ax.set_zlim([0, 10])
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('v1, v2, v3')

# 添加图例
ax.legend()

# 显示图形
plt.show()
```
