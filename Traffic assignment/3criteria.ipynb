```julia
using Plots
```

```julia
function dominant_area(path_attr)
    dim = 300
    path_choice = zeros(dim, dim)
    for i in 1:dim
        for j in 1:dim
            path_cost = 0.01*i * path_attr[:,1] + path_attr[:,2] + 0.01*j * path_attr[:,3]
            path_choice[i, j] = argmin(path_cost)
        end
    end
    heatmap(path_choice, xlabel="VOT", ylabel="VOD")
end
```

```julia
# time, toll, distance
path_attr = Float64[
    12 10 12;
    12 12 10;
    10 12 12;
    ]
dominant_area(path_attr)
```
