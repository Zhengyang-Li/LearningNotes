```julia
using Plots
```

```julia
function dominant_area(path_attr; lb=0, ub=3)
    Δ = 0.01
    dim = Int((ub - lb) / Δ)
    path_choice = zeros(dim, dim)
    for i in 1:dim
        for j in 1:dim
            vot = lb + Δ*i
            vor = lb + Δ*j
            path_cost = vot * path_attr[:,1] + path_attr[:,2] + vor * path_attr[:,3]
            path_choice[i, j] = argmin(path_cost)
        end
    end
    heatmap(path_choice, xlabel="VOR", ylabel="VOT")
end
```

```julia
function path_cost(path_attr, VOT, VOR)
    return VOT * path_attr[:,1] + path_attr[:,2] + VOR * path_attr[:,3]
end
```

```julia
# time, toll, distance
path_attr = Float64[
    12 10 12;
    12 12 10;
    10 12 12;
    ]
dominant_area(path_attr)
```
